### Relationship between Previous Cancellations and current Cancellations

```python
# Finding the correlation between is_canceled and previous_cancellations
df[['is_canceled','previous_cancellations']].corr() * 100
```

```python
# Sort previous cancellations by index (cancellation times)
df['previous_cancellations'].value_counts().sort_index()
```

```python
# Let's take a look customer cancellation rate for customers who cancelled more than 10 times before!
df[df['previous_cancellations'] > 10]['is_canceled'].mean() * 100
```

```python
# Convert index to a list
a = df['previous_cancellations'].value_counts().index.to_list()
# Sort the list by ascending order
a.sort()
a
```

```python
# Finding the current cancellation rate according to their previous cancellation
b = []
for ccltime in a:
    b.append(round(df[df['previous_cancellations']==ccltime]['is_canceled'].mean() * 100 , 2))
print(b)
```

```python
# Creating a dictionary with keys as previous_cancellation and values as current_cancellation_rate
cancellation_rate = {'previous_cancellation': a, 'current_cancellation_rate': b}
cancellation_rate_data = pd.DataFrame(cancellation_rate)
cancellation_rate_data
```

```python
# Plotting the graph depicting Correlation between Current Cancellation Rate and Previous Cancellation using seaborn
plt.figure(figsize = (12,6), dpi=110)
sns.lineplot(x = 'previous_cancellation', y = 'current_cancellation_rate', data = cancellation_rate_data)
plt.grid()
plt.title('Correlation between Current Cancellation Rate and Previous Cancellation', weight='bold')
plt.xlabel('Previous Cancellations')
plt.ylabel('Current Cancellation Rate (%)')
```

*   According to graph, when customer previouslly had one time hotel reservation cancellation, 94.58% of them cancelled again. 
*   Cancellation rate for customers who have canceled more than 10 times before: 85.56%

#### Overview of reserved room type and assigned room type

```python

# Selecting three colums from DataFrame which are: "is_canceled", "reserved_room_type", "assigned_room_type"
df[['is_canceled','reserved_room_type','assigned_room_type']]
```

```python
# Define a function to check whether a customer get their preferred reserved room type
def rrt_art_same(df):
  '''
  This function returns whether the customer get their preferred reserved room type or not.
  '''
  if df['reserved_room_type'] == df['assigned_room_type']:
      return 'True'
  else:
      return 'False'
```

```python
# Apply to dataframe
df['rrt_art_same'] = df.apply(rrt_art_same,axis=1)

# Check dataframe
df[['is_canceled','reserved_room_type','assigned_room_type','rrt_art_same']]
```

Type of customer booking:

2.   **Group** — when the booking is associated to a group
3.   **Transient** — when the booking is not part of a group or contract, and is not associated to other transient booking
4.   **Transient-party** — when the booking is transient, but is associated to at least other transient booking

Transient type of customer is the main source of booking because more than 75% of booking coming from this side after that Transient-Party, Contract and Group are coming in the focus.

### Feature Selection and Feature Engineering

```python
# Making a copy of the DataFrame
df_subset = df.copy()
```

```python
# Dropping the unnecessary rows
df_subset = df_subset.drop(df_subset[(df_subset.adults+df_subset.babies+df_subset.children)==0].index)
```

```python
# Making a new column "Room" with value 0
df_subset['Room'] = 0
```

```python
# Assigning the value where "reserved_room_type" is same as "assigned_room_type" with 1 in "Room" Column
df_subset.loc[df_subset['reserved_room_type'] == df_subset['assigned_room_type'],'Room'] = 1
```

```python
# Making a new column "net_cancelled" with value 0
df_subset['net_cancelled'] = 0
```

```python
# Assigning the value where "previous_cancellations" is greater than "previous_bookings_not_canceled" with 1 in "net_cancelled" Column
df_subset.loc[df_subset['previous_cancellations'] > df_subset['previous_bookings_not_canceled'],'net_cancelled'] = 1
```

```python
# Dropping the unnecessary categorical columns
df_subset = df_subset.drop(['arrival_date_year','arrival_date_week_number','arrival_date_day_of_month',
                            'arrival_date_month','assigned_room_type','reserved_room_type','reservation_status_date',
                            'previous_cancellations','previous_bookings_not_canceled','reservation_status'],axis=1)
```
